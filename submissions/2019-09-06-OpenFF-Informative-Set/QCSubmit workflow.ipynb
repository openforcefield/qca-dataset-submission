Notebook to show the generation of the informative optimization dataset starting with the filtered molecules.

```python
from qcsubmit.factories import OptimizationDatasetFactory
from qcsubmit import workflow_components
```

Build the factory and add a conformer generator component.

```python
opt_factory = OptimizationDatasetFactory()
# as agreed low priority
opt_factory.priority = "low"
opt_factory
```

```python
# add the conformer generator
conformer_generator = workflow_components.StandardConformerGenerator()
conformer_generator.max_conformers = 40
opt_factory.add_workflow_component(conformer_generator)
```

```python
# write out the factory settings
opt_factory.export_settings("informative_settings.yaml")
```

```python
# now process the molecules
dataset = opt_factory.create_dataset(dataset_name="Ehrman Informative Optimization v1.0", molecules="selected.smi", description="This provides an optimization dataset based on an initial batch of Jordan Ehrman's analysis of eMolecules, pulling out molecules with minimized geometries which are substantially different in different force fields.", tagline="Geometry optimization of informative molecules.")
```

```python
# set the url for where the info will be stored
dataset.metadata.long_description_url = "https://github.com/openforcefield/qca-dataset-submission/tree/master/2019-09-06-OpenFF-Informative-Set"
```

```python
dataset.export_dataset("informative_optimization_dataset.json")
```
