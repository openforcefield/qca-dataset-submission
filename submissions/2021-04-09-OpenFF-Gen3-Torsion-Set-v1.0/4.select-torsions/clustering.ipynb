```python
import pickle
tid_molecules_list =pickle.load(open('tid_molecules_list.p','rb'))
```

```python
# select core, choose around 10 entries per torsion parameter
# resulting tid_clusters_list[tid] = [ {'cluster_label': N, 'torsions': [...]}]
tid_clusters_list = gen_tid_clusters_list(tid_molecules_list, fptype=OEFPType_MACCS166, lim=10, select_option =1) 
```

```python
print(f'avg: {sum([len(clusters) for tid, clusters in tid_clusters_list.items() if len(clusters)>0])/sum([1 for tid, clusters in tid_clusters_list.items() if len(clusters)>0]):.2f}')
print(f'max: {max([len(clusters) for tid, clusters in tid_clusters_list.items()])}')
print(f'min: {min([len(clusters) for tid, clusters in tid_clusters_list.items() if len(clusters)>0])}')
```

```python
# duplicate issue
count = 0
tot = 0
for tid, clusters in tid_clusters_list.items():
    if len(clusters) == 0:
        # print(tid)
        count += 1
    tot += len(clusters)
print(f'# uncovered: {count} (out of {len(tid_clusters_list)})')
print(f'# entries: {tot}')
```
