```python
from openforcefield.topology import Molecule
```

```python
mol1 = Molecule.from_smiles('[H]c1c(c(c(c(c1[H])[H])[S+2](N([H])[H])([O-])[O-])[H])[H]')
mol1.visualize()
```

```python
mol2 = Molecule.from_smiles('[H]c1c(c(c(c(c1[H])[H])S(=O)(=O)N([H])[H])[H])[H]')
mol2.visualize()
```

```python
mol1.is_isomorphic_with(mol2)
```

```python
inchi1 = mol1.to_inchikey(fixed_hydrogens=True)
inchi1
```

```python
inchi2 = mol2.to_inchikey(fixed_hydrogens=True)
inchi2
```

```python
inchi1 == inchi2
```
